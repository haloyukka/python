# 読み込んだデータをソートする
平成28年度の人口データを元に人口の少ない順に並び替えて表示してみよう。
並び替え(ソート)を行うには、sort_valuesメソッドを使う。
ascending=Trueの部分を、ascending=Falseと書き換えると、人口の多い順(つまり、降順)に並び替えてくれる。

```python
import pandas as pd
df = pd.read_csv("population.csv",encoding="SHIFT-JIS")
df.sort_values(by=["平成28年"],ascending=True)
```

## 棒グラフを描画する
