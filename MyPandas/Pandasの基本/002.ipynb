# Group byによる集計
NaNデータは除外される。

```python
import pandas as pd

df = pd.read_excel('data.xlsx')
display(df)
```

```python
df.groupby('曜日').mean()
```

```python
df.groupby('曜日').max(numeric_only=True)
```

```python
df.groupby('曜日').count()
```

# index・columns
インデックスの変更<br>
カラム名の変更<br>
ある列をインデックスに設定<br>



```python
import pandas as pd

df = pd.read_excel('data.xlsx')
display(df)
```

```python
[str(x).zfill(3) for x in range(1,10)]
```

```python
df.index = [str(x).zfill(3) for x in range(1,10)]
```

```python
display(df)
```

```python
df.columns = ['date', 'day of week', 'name', 'sales']
```

```python
display(df)
```

```python
df.set_index('date', inplace=True) #inplaceを使用するのでDataFrameが書き換わる
```

```python
display(df)
```

```python
df.reset_index(inplace=True)
```

```python
display(df)
```

# merge結合
concatとの違いは、あるキーを使って結合させる点。<br>
3つ以上同時に結合することができない。

```python
import pandas as pd

df_1 = pd.read_excel('merge1.xlsx')
df_2 = pd.read_excel('merge2.xlsx')
display(df_1)
display(df_2)
```

```python
pd.merge(df_1,df_2, on='ユーザID')
```

# DataFrameのmap
map + lambda式

```python
import pandas as pd
df = pd.read_excel('merge1.xlsx')
display(df)
```

```python
df['年齢'].map(lambda x: '成人' if x >= 20 else '未成年')
```

```python
df['区分'] = df['年齢'].map(lambda x: '成人' if x >= 20 else '未成年')
```

```python
display(df)
```
